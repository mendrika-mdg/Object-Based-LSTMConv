# Number of cores per context domain for June-September (2004-2024)

This is the number of cores every time steps (time resolution is 15 minutes) from 20 years of data

```python
import numpy as np                                                                      
import matplotlib.pyplot as plt                                                         
```

```python
nb_cores = np.load("/home/users/mendrika/Object-Based-LSTMConv/outputs/nbx0/nbx0-ws.npy")
```

```python
plt.figure(figsize=(7, 5))
plt.hist(nb_cores, bins=20, color="green", alpha=0.7, edgecolor='black')
plt.axvline(x=3, color='red', linestyle='--', linewidth=2, label=f'Mean={round(np.mean(nb_cores))}')
plt.xlabel("Number of Cores", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
