# Original

```python
original_example = torch.load("/work/scratch-nopw2/mendrika/OB/raw/shards/t0/train_t0/shard_112.pt")
```

```python
X = original_example["X"]
```

```python
X.shape
```

```python
scaler = torch.load("/home/users/mendrika/Object-Based-LSTMConv/outputs/scaler/scaler_realcores.pt", weights_only=False)
mean = np.asarray(scaler["mean"])
scale = np.asarray(scaler["scale"])

MASK_COL_INDEX = 12
COLS_TO_SCALE = range(4, 12)
```

```python
flat = X.numpy().reshape(-1, X.shape[-1])
rows = np.where(flat[:, MASK_COL_INDEX] == 1)[0]

scaled = (flat[np.ix_(rows, COLS_TO_SCALE)] - mean) / scale
flat[np.ix_(rows, COLS_TO_SCALE)] = scaled.astype(np.float32)
```

```python
X_scaled = flat.reshape(X.shape)
X_scaled = torch.tensor(X_scaled, dtype=torch.float32)
```
